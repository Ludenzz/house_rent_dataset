```python
#Mean, Median, Maximum and Minimum House Rents

print("Highest House Rent: {}".format(round(df["Rent"].max())))
print("Lowest House Rent: {}".format(round(df["Rent"].min())))
print("Mean House Rent: {}".format(round(df["Rent"].mean())))
print("Median House Rent: {}".format(round(df["Rent"].median())))
```

```python
#corr

df.corr()
```

```python
fig, ax = plt.subplots(figsize = (18, 18))
sns.heatmap(df.corr(), annot = True, linewidths = .5, fmt = '.2f', annot_kws={
                "fontsize": 40,
                "fontweight": "bold",
                "fontfamily": "serif"
            })
plt.show()
```

```python
#Heatmap on BHK vs Area Type

plt.figure(figsize = (20, 6))
colormap = plt.cm.plasma
sns.heatmap(pd.crosstab(df["Area Type"], df["BHK"]), cmap = colormap)
```

```python
#Number of House in Each City which is Available for Rent
city_counts = df["City"].value_counts()

city_counts
```

```python
#Bar Plot for Number of House in Each City which is Available for Rent

f, ax = plt.subplots(figsize = (20, 10))
ax = city_counts.plot(kind = "bar", color = "blue", rot = 0)

for container in ax.containers:
    ax.bar_label(container)
```

```python
#Numbers of Furnishing Status

fur_stat_counts = df["Furnishing Status"].value_counts()
fur_stat_counts
```

```python
#Distribution of Furnishing Status, visualization

f, ax = plt.subplots(figsize = (20, 10))
ax = fur_stat_counts.plot(kind = 'bar',color = "green", rot = 0)

for container in ax.containers:
    ax.bar_label(container)
```

```python
# Distribution of Tenant Preferred

ten_pref_counts = df["Tenant Preferred"].value_counts()
ten_pref_counts
```

```python
# Distribution of Tenant Preferred, visualization

f, ax = plt.subplots(figsize = (20, 10))
ten_pref_counts.plot(kind = "bar", color = "orange", rot = 0)

for container in ax.containers:
    ax.bar_label(container)
```

```python
#Numbers of BHK

bhk_counts = df.BHK.value_counts()
bhk_counts
```

```python
#Distribution of BHK, visualization

f, ax = plt.subplots(figsize = (20, 10))
bhk_counts.plot(kind = "bar", color = "grey")

for container in ax.containers:
    ax.bar_label(container)
```

```python
#Pie Plot on Cities to check the distribution

plt.figure(figsize = (20, 8))
myexplode = (0.1, 0, 0, 0, 0, 0)

city_counts.plot(kind = "pie", explode = myexplode, shadow = True, autopct = "%1.1f%%")

plt.axis("equal")
# plt.legend(labels = city_counts.index)
plt.show()
```

```python
#Scatter Plot on House Sizes vs House Rent

plt.figure(figsize = (20, 6))
plt.ticklabel_format(style = "plain")
plt.scatter(df.Size, df.Rent, alpha = .5, color = "red")
plt.xlabel("Size")
plt.ylabel("Rent")
plt.show()
```

```python
#Relationship between House Sizes vs House Rent

plt.figure(figsize = (25, 8))
plt.ticklabel_format(style = "plain")
plt.scatter(df.Size, df.Rent, c = df["Size"], s = df["Size"], alpha = 0.5, cmap = 'viridis')
plt.xlabel("Size")
plt.ylabel("Rent")
plt.colorbar()
```

```python
#Bar Plot for City vs House Rent

plt.figure(figsize = (20, 6))
sns.barplot(x = df["City"], y = df["Rent"])
plt.show()
```

```python
#Histogram on House Sizes

plt.figure(figsize = (20, 6))
df.Size.plot(kind = "hist", bins = 20)
plt.show()
```

```python
#Histogram on House Sizes - other option

plt.figure(figsize = (20, 6))
df.Size.hist(bins = 20)
```
